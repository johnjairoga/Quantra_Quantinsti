## Read the Data

The stock price of Bank of America and J.P. Morgan, from January 2020 to March 2021, is stored in a CSV file `jpm_and_bac_price.csv`. Along with the close prices, the file also contains the daily returns of Bank of America. You can use `pandas.read_csv` to read the CSV file. 

````python
# Import pandas
import pandas as pd

# The data is stored in the directory 'data_modules'
path = '../data_modules/'

# Read the data
data = pd.read_csv(path + 'jpm_and_bac_price.csv', index_col=0)

# Convert the index to datetime
data.index = pd.to_datetime(data.index)

# Display the data
data.head()
````

<a id='line-graph'></a> 
## Line Graph of Close Price

To plot different graphs, you will use the `matplotlib` Python library. The syntax to import the library and set the style is shown below:

Syntax: 
```python
import matplotlib.pyplot as plt
%matplotlib inline
style.use(<style-name>)
```

* `%matplotlib inline` is used to render the graphs in the notebook. Without this, you will not be able to see graphs.
* `style.use(<style-name>)` is used to set the style of the graph to suit your requirement. You can check all the available styles using `plt.style.available`.

After you have imported the library and set the style, you can plot the line graph using the pandas' `plot` method.

Syntax:
```python
dataframe.column.plot(figsize, color)
```
1. **figsize**: Set size of x and y-axis
2. **colour**: Set colour for line


Once you plot the graph, you can add the title and axis labels using the `title`, `xlabel` and `ylabel` methods respectively.

Syntax:
```python
plt.title(title, fontsize)
plt.xlabel(x-label, fontsize)
plt.xlabel(y-label, fontsize)
```

1. **title:**  Title in string format
2. **x-label:**  X-label in string format
3. **y-label:**  Y-label in string format
4. **fontsize:**  Size of the font

````python
# Import library for data visualisation
import matplotlib.pyplot as plt
%matplotlib inline

# Set the plot style
plt.style.use('seaborn-darkgrid')

# Plotting the close price column
data['Close_BAC'].plot(figsize=(10, 7), color='green')
# Setting the graph title
plt.title("Closing Price of BAC", fontsize=14)
# Setting the x-axis title
plt.xlabel('Year-Month', fontsize=12)
# Setting the y-axis title
plt.ylabel('Price ($)', fontsize=12)
# Function for showing graph
plt.show()
````

<a id='scatter-plot'></a> 
## Scatter Plot of Two Stocks' Close Price

Scatter plot is used to study the relationship between two variables. To plot the scatter plot of two columns in a dataframe, you will use the `scatter` method of `matplotlib`. The syntax is shown below:

Syntax: 
```python
import matplotlib.pyplot as plt
plt.scatter(col1, col2)
```
1. **col1**: First column (x-axis)
2. **col2**: Second column (y-axis)

````python
# Plot of BAC vs JPM
plt.figure(figsize=(10,7))
plt.scatter(data['Close_BAC'], data['Close_JPM'])

# Set the title of the plot
plt.title('BAC Price Vs. JPM Price', fontsize=14)
# Setting the x-axis title
plt.xlabel('BAC Price', fontsize=12)
# Setting the y-axis title
plt.ylabel('JPM Price', fontsize=12)
# Function for showing graph
plt.show()
````

<a id='daily-returns'></a> 
## Histogram of Daily Returns

To plot the histogram of a column in a dataframe, you will use the `plot` method of `pandas`. The syntax is shown below:

Syntax: 
```python
DataFrame[col].plot(kind='hist')
```
1. **col**: Column name
2. **kind**: Kind of plot (in case of histogram, it is `'hist'`)

````python
# Histogram of daily returns
plt.figure(figsize=(10,7))
data['daily_returns_BAC'].plot(kind='hist')

# Set the title of the plot
plt.title('Daily Returns of BAC', fontsize=14)
# Setting the x-axis title
plt.xlabel('Returns', fontsize=12)
# Setting the y-axis title
plt.ylabel('Frequency', fontsize=12)
# Function for showing graph
plt.show()
````
